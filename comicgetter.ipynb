```python
ComicURL = "https://mycomiclist.com/brzrkr/issue-8/256514/all"
ComicName = "bzrkr#8"
```

```python
import requests
import json
```

```python
r = requests.get(ComicURL)
```

```python
# find all img tags with src that looks like "2.bp.blogspot.com/..."
# and extract the src attribute
import re
exp = re.compile(r'https://2.bp.blogspot.com/[^"]+"')
# apply the regex to the response
imgs = exp.findall(r.text)
```

```python
# download all imgs into a folder, and make a pdf
import os
import urllib.request
if not os.path.exists(ComicName):
    os.mkdir(ComicName)
# download all images using wget
for idx,img in enumerate(imgs):
    os.system("wget "+ img[:-1] + " -O "+ComicName+"/img"+str(idx)+".jpg")
```

```python
os.system("convert -delay 10 -loop 0 {}/*.jpg {}.pdf".format(ComicName,ComicName))
```
